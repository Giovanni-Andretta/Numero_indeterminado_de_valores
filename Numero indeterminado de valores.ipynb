```python
#Giovanni Andretta Carbonero

#Lista
notas = []

#Função para calcular quantas notas existem abaixo de 7 na lista notas
def abaixo_da_sete():
    
    count = 0
    
    for i in notas:
        
        if i < 7:
            count+=1
            
    return count

#Função para calcular quantas notas existem acima da media na lista notas
def acima_da_media():
    
    count = 0
    
    for i in notas:
        
        if i > media():
            count+=1
    
    return count
         
#Função para calcular a media somando todas as notas da lista e dividindo pela quantidade de valores
def media():
    
    media = sum(notas)
    media = media / len(notas)
    
    return media


#Função para exibir as infomações
def exibe():
    
    print("="*50 + "\n\n")
    
    print("Quantidade de valores que foram lidos: " + str(len(notas)))
    print("\nValores na ordem em que foram informados: " + str(notas))
    
    notas.reverse()
    print("\nValores na ordem inversa à que foram informados: " + str(notas))
    notas.reverse()
    
    print("\nSoma dos valores: " + str(sum(notas)))
    
    print("\nMédia dos valores: " + str(media()))
    
    print("\nQuantidade de valores acima da média: " + str(acima_da_media()))
    
    print("\nQuantidade de valores abaixo de sete: " + str(abaixo_da_sete()))
    
    print("\n\nPrograma executado com sucesso!\n\n\tFechando programa...")
    
    print("\n\n" + "="*50)

#Função para receber as informações de entrada referente as notas, e coloca-las dentro da lista notas
def recebe():
    
    opt = 0
    
    while opt != -1:
        
        print("\n" + "="*50 + "\n\n")
        
        nota = float(input("Insira a nota: "))
        
        if nota < 0:
            print("\n\nValor invalido...\nTente novamente\n\n")
            opt = 1
        else:
            notas.append(nota)
        
        opt = int(input("\nDeseja adicionar outra nota?\n\nDigite qualquer numero para sim e -1 para não\n\nDigite: "))

#Função principal onde chama as principais funções
def main():
    
    recebe()
    exibe()
    
main()
```
